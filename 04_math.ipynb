```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('AmesHousing.csv')
```

1 - Computing the mean and standard deviation.

```python
y_array = df['SalePrice'].to_numpy() 

manual_mean = np.mean(y_array)
manual_std = np.std(y_array)

print(f"Mean: {manual_mean}")
print(f"Std: {manual_std}")
# I converted the target to a NumPy array to calculate the statistics manually. This follows the course requirement to use NumPy's core functions (.mean and .std) instead of the built in Pandas methods.
```

2 - Standardise one column by hand using broadcasting.

```python
from sklearn.preprocessing import StandardScaler
X_raw = df["Gr Liv Area"].to_numpy() # I chose 'Gr Liv Area' because more living space directly increases the price, as shown in our scatter plot.

mean_val = np.mean(X_raw)
std_val = np.std(X_raw)

# Applying the Z-score formula manually using Broadcasting.
z_manual = (X_raw - mean_val) / std_val
# I returnd to the course slides to find the right formula.(page:285,day:39) 

scaler = StandardScaler()
z_scaler = scaler.fit_transform(X_raw.reshape(-1, 1)).flatten()
if np.allclose(z_manual, z_scaler):
    print("Correct!")
else:
    print("There is a difference.")
# I calculated the Z-score manually to show how we normalize data so that different features are on the same scale for the model.
```

3 - Computing cosine similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
# Assisted by AI to identify the specific row labels for max/min prices and the library.
max_price = df["SalePrice"].max()
min_price = df["SalePrice"].min()

highest = df[df["SalePrice"] == max_price].iloc[0] # Using .iloc[0] ensures I'm working with a single record (series) for each.
lowest = df[df["SalePrice"] == min_price].iloc[0]

num_cols = df.select_dtypes(include=["number"]).columns # I used select_dtypes to automatically identify all numerical columns in the dataset. 
high_vec = highest[num_cols].fillna(0)
low_vec = lowest[num_cols].fillna(0)

similarity = cosine_similarity([high_vec], [low_vec]) # Measuring the feature-wise similarity between the two extreme cases.
print(similarity)
```

4 - Estimate a probability.

```python
threshold = df["SalePrice"].median()

# Identifying high quality elements (based on words if they are texts).
high_quality = df[df["Exter Qual"].isin(["Ex", "Gd"])]

# Calculating the probability (the proportion of those whose price is above average).
prob = (high_quality["SalePrice"] > threshold).mean()
print(f"Probability: {prob:.4f}")
```
